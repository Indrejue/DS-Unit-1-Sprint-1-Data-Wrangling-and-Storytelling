```python
#load extracted csv into a pandas dataframe to work with data
#header to define the correct row with header information
#skipfooter to remove extra data at the foot of the file
loandf=pd.read_csv('LoanStats_2018Q4.csv', header=1,skipfooter=2)
```

```python
#check the first few rows of DF to see it loaded properly
loandf.head()
```

```python
#check bottom of file to make sure extra end data was removed
loandf.tail()
```

```python
#check shape of DF to see how many values for each item
loandf.shape
```

```python
#check on colums that are entirelly NaN to remove
loandf.isnull().sum().sort_values(ascending=False)
```

```python
#remove unnecisary columns from DF
loandf=loandf.drop(['id','member_id','url','desc'],axis=1)
loandf.head()
```

```python
#prepare a function to remove information causing intrate to be a string instead of a float value
def int_to_float (intrate):
  return float(intrate.strip().strip('%'))
```

```python
#pass through function to int_rate to turn column into float values
loandf['int_rate']=loandf['int_rate'].apply(int_to_float)
loandf.head()
```

```python
#confirm data is float
loandf['int_rate'].dtype
```

```python
#define a function to make emp_title column more uniform in the way it presents data
def title_clean(emp_title):
  if isinstance(emp_title,str):
    return emp_title.strip().title()
  else:
    return 'Unknown'
```

```python
#apply defined function
loandf['emp_title']=loandf['emp_title'].apply(title_clean)
loandf.head(10)
```

```python
#turn issue date into a number format that can be worked with in functions
loandf['issue_d']=pd.to_datetime(loandf['issue_d'], infer_datetime_format=True)
```

```python
#turn earliest cr line into a numarical format for funtions
loandf['earliest_cr_line']=pd.to_datetime(loandf['earliest_cr_line'], infer_datetime_format=True)
```

```python
#define function ro turn the term data into an int value for use in functions
def termint(termtime):
  return int(termtime.strip(' months'))
```

```python
#apply function to turn term into integer data
loandf['term']=loandf['term'].apply(termint)
loandf.head()
```

```python
#define function to return value baased on loan status
def loanstat(loan_status):
  if loan_status=='Fully Paid':
    return 1
  elif loan_status=='Current':
    return 1
  else:
    return 0
```

```python
#apply function to loan status column to generate a new column
loandf['loan_status_great']=loandf['loan_status'].apply(loanstat)
loandf.head()
```

```python
#define function to load end year
loandf['last_pymnt_year'] = loandf['term']/12 + loandf['issue_d'].dt.year
loandf.head()
```

```python
#make last payment date into date formate
loandf['last_pymnt_d'] = pd.to_datetime(loandf['last_pymnt_d'],infer_datetime_format=True)
#function for last payments made year
loandf['last_pymnt_d_year'] = loandf['last_pymnt_d'].dt.year
loandf.head()
```

```python
#function for last payment made month
loandf['last_pymnt_d_month']=loandf['last_pymnt_d'].dt.month
loandf.head()
```

```python
#define function to load end month of loan
loandf['last_pymnt_month'] = (loandf['term'] + loandf['issue_d'].dt.month)%12
loandf.head()
```

```python
#replace all the non remainder values with 12 to represent the month of dec
loandf['last_pymnt_month']=loandf['last_pymnt_month'].replace({0:12})
```

```python
#check that we have different months for end date
loandf['last_pymnt_month'].value_counts()
```
